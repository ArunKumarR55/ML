```python
import numpy as np

training_inputs = np.array([[1,1],[-1,1],[1,-1],[-1,-1]])
expected_output = np.array([1,-1,-1,-1])

weights = np.zeros(2)
bias = 0
learning_rate =0.1
epochs = 10

for epoch in range(epochs):
    for i in range(len(training_inputs)):
        predicted_output = np.dot(training_inputs[i],weights) + bias
        error = expected_output[i] - predicted_output
        weights = weights + learning_rate * error * training_inputs[i]
        bias = bias + learning_rate * error
print("Final Weights",weights)
print("Final Bias",bias)
```
